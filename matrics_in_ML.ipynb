# Confusion matrix:

### 1. True Positive(TP): When passenger is predicted as a +ve when it is actually +ve 

### 2. False Positive(FP): When passenger is predicted as -ve when it is +ve

### 3.False Negative(FN): When passenger is predicted -ve when it is +ve

### 4. True Negative(TN):  When passenger is predicted +ve when it is -ve

Actual values
TP      FP
FN      TN      




## Type 1 Error(FPR- False Positive Rate)

FPR= FP/FP+TN

## Type 2 Error(FNR-False Negative Rate)

FNR=FN/FN+TP

## Recall(TPR-True Positive Rate)(Sensitivity):
Out of total +ve values how many +ve values are predicted correctly

recall=TP/TP+FN

## Precision-(Positive Predicted Values)(Specificity):
Out of total +ve predicted values how many are actually +ve

precision = TP/TP+FP

**Note: When FP is imp then go with precision, and when FN is imp go with recall**

## F-Beta: This is used to combine precision and recall

FBeta = (1+B)^2*((Precision*recall)/(B^2)*(Precision + recall))

**Note: If FN and FP both are equally important then we use B=1,it will be know F1 Error**
 if B=>1-10 then select FN
 if B=>0-1 then select FP
 
 
## Accuracy:
   accuracy= TP+TN/(TP+FP+FN+TN)
